## ***13. Find Total Number of Persons Having Age Between 17 To 48 (Inclusive) Using Between Method And Also Without Using Between Method***

```python
count_between = df['age'].between(17, 48).sum()

print("Total number of persons with age between 17 and 48 (inclusive) using between method:", count_between)

count_without_between = df[(df['age'] >= 17) & (df['age'] <= 48)].shape[0]

print("Total number of persons with age between 17 and 48 (inclusive) without using between method:", count_without_between)
```

## ***17. Replace income value [<=50k with 0 and >50k with 1]***

```python
df['income'] = df['income'].replace({'<=50K': 0, '>50K': 1})

df
```

## ***18. Which Workclass getting The Highest Salary***

```python
average_salary_by_workclass = df.groupby('workclass')['income'].mean()

highest_salary_workclass = average_salary_by_workclass.idxmax()
highest_salary = average_salary_by_workclass.max()

print(f"The workclass getting the highest salary is '{highest_salary_workclass}' with an average salary of {highest_salary:.2f}.")
```

## ***19. Who Has Better Chance To Get Salary >50k Male or Female***

```python

male_count = df[(df['gender'] == 'Male') & (df['income'] == 1)].shape[0]
female_count = df[(df['gender'] == 'Female') & (df['income'] == 1)].shape[0]

total_males = df[df['gender'] == 'Male'].shape[0]
total_females = df[df['gender'] == 'Female'].shape[0]


male_chance = male_count / total_males * 100 if total_males > 0 else 0
female_chance = female_count / total_females * 100 if total_females > 0 else 0

print(f"Males have a {male_chance:.2f}% chance to earn >50k.")
print(f"Females have a {female_chance:.2f}% chance to earn >50k.")

better_chance = 'Males' if male_chance > female_chance else 'Females' if female_chance > male_chance else 'Equal'
print(f"{better_chance} have a better chance of earning >50k.")
```
